```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd

ReviewData = pd.read_excel('C:/Users/Sneha-laptop/Anaconda2/Dow-Project/wipes_reviews.xlsx',
                     sheetname = 'wipes_reviews_test',
                     header = 0,
                     index_col = 0,
                     parse_cols = "B,E")


Dataformat=list(ReviewData.values.flatten())

sid = SentimentIntensityAnalyzer()
count=0
for sentence in Dataformat:
        count=count+1
        print(sentence)
        ss = sid.polarity_scores(sentence)
        for k in sorted(ss):
            print('{0}: {1}, '.format(k, ss[k]))
        print()

print count
```
